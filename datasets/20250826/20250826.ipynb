```python
import polars as pl

raw_data = pl.scan_csv("billboard.csv", infer_schema_length=None, null_values='NA')

print(raw_data.collect_schema().names())
```

```python
raw_data.describe()
```

```python

data = raw_data.with_columns(pl.col("date").str.to_datetime("%Y-%m-%dT%H:%M:%SZ"))\
.with_columns(pl.col("date").cast(pl.Date))

print(data.collect_schema().names())
```
